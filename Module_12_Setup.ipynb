### Part 2: Update Database

```python
# Create a dictionary for the new restaurant data
Penang_Flavours= {
        "BusinessName":"Penang Flavours",
        "BusinessType":"Restaurant/Cafe/Canteen",
        "BusinessTypeID":"",
        "AddressLine1":"Penang Flavours",
        "AddressLine2":"146A Plumstead Rd",
        "AddressLine3":"London",
        "AddressLine4":"",
        "PostCode":"SE18 7DY",
        "Phone":"",
        "LocalAuthorityCode":"511",
        "LocalAuthorityName":"Greenwich",
        "LocalAuthorityWebSite":"http://www.royalgreenwich.gov.uk",
        "LocalAuthorityEmailAddress":"health@royalgreenwich.gov.uk",
        "scores":{
            "Hygiene":"",
            "Structural":"",
            "ConfidenceInManagement":""
        },
        "SchemeType":"FHRS",
        "geocode":{
            "longitude":"0.08384000",
            "latitude":"51.49014200"
        },
        "RightToReply":"",
        "Distance":4623.9723280747176,
        "NewRatingPending":True
    }
```

```python
# Insert the new restaurant into the collection
establishments.insert_one(Penang_Flavours)
```

```python
# Check that the new restaurant was inserted
query = {'BusinessName': 'Penang Flavours'}
results = establishments.find(query)

# print 
for result in results:
    pprint(result)
```

```python
# Find the BusinessTypeID for "Restaurant/Cafe/Canteen" and return only the BusinessTypeID and BusinessType fields
# to see what the BusinessTypeID for our new restuarant as it is blank in the dict
# parameters
query = {"BusinessType": "Restaurant/Cafe/Canteen"}
feilds = {"BusinessTypeID": 1, "BusinessType": 1, "_id": 0}

# set results 
results = establishments.find(query, feilds)

# print
for result in results:
    print(result)
```

```python
# Update the new restaurant with the correct BusinessTypeID
# parameters
query = {'BusinessName': 'Penang Flavours'}
update = {"$set": {"BusinessTypeID": "1"}}

# set results 
result = establishments.update_one(query, update)
```

```python
# Confirm that the new restaurant was updated
query = {'BusinessName': 'Penang Flavours'}
fields = {'BusinessName': 1, 'BussinessType': 1, 'BusinessTypeID': 1, '_id': 0}

# set results 
results = establishments.find(query, fields)

# print 
for result in results:
    pprint(result)
```

```python
# Find how many documents have LocalAuthorityName as "Dover"
# parameters 
query = {'LocalAuthorityName': 'Dover'}
count_of_Dover = establishments.count_documents(query)

# print
print(count_of_Dover)
```

```python
# Delete all documents where LocalAuthorityName is "Dover"
delete_Dover = establishments.delete_many({'LocalAuthorityName': 'Dover'})
```

```python
# for all documents change the data type from String to Decimal for longitude and latitude
# parameters
convert_latlon = [
    {
        "$set": {
            "geocode.latitude": {"$toDouble": "$geocode.latitude"},
            "geocode.longitude": {"$toDouble": "$geocode.longitude"}
        }
    }
]
# set results
result = establishments.update_many({}, convert_latlon)
```

```python
# convert RatingValue to interger numbers 
# Set non 1-5 Rating Values to Null
non_ratings = ["AwaitingInspection", "Awaiting Inspection", "AwaitingPublication", "Pass", "Exempt"]
establishments.update_many({"RatingValue": {"$in": non_ratings}}, [ {'$set':{ "RatingValue" : None}} ])

# parameters
convert_ratings = [
    {
        '$set': {'RatingValue': {'$toInt': '$RatingValue'}}
    }
]

# set results 
result = establishments.update_many({}, convert_ratings)
```
